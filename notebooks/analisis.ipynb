### Tipos de datos
- name: Nombre del repositorio (Categorico Nominal)
- full_name: Nombre completo del repositorio (Categorico Nominal)
- description: Descripcion del repositorio (Categorico nominal)
- stars: Cantidad de estrellas del repositorio (Numerico continuo)
- forks: Cantidad de forks al repsoritorio (Numerico continuo)
- watchers: Cantidad de visualizaciones (Numerico continuo)
- languaje: Lenguaje principal del proyecto
- created_at: Fecha de creacion (Fecha)
- updated_at: Fecha de ultima actualizacion (fecha)
- open_issues:
- license: Tipo de licencia del repositorio
- topics: Topicos del repositorio
- size_kb: Tamano del repositorio (Numerico continuo)
- is_fork: Es forkeable el proyecto (Categorico binario)
- archived: Repositorio archivado (Categorico binario)
- homepage: direccion de los creadores del repositorio

```python
github_data.head()
```

```python
# distribucion de estrellas
sns.histplot(github_data['stars'])
plt.title("Distribucion de estrellas")
plt.show()
```

```python
# distribuciones de forks
sns.histplot(github_data['forks'])
plt.title("Distribucion de forks")
plt.show()
```

```python
# distribucion de visualizaciones
sns.histplot(github_data['watchers'])
plt.title("Distribucion de visualizaciones")
plt.show()
```

```python
# distribuciones de tamaños en kb de repositorios
sns.histplot(github_data['size_kb'])
plt.title('Distribucion de tamano en kb de')
plt.show()
```

```python
# lenguajes mas utilizados dentro de los datos
plt.figure(figsize=(12, 5))
# grafico de barras
sns.countplot(data=github_data, x='language')
plt.title("Lenguajes mas utilizados")
plt.xticks(rotation=45)
plt.show()
```

Los lenguajes mas utilizados son:
1. Python
2. TypeScript
3. JavaScript

```python
# tipos de licencias mas utilizadas para los repositorios
plt.figure(figsize=(12, 5))
sns.countplot(data=github_data, x='license')
plt.title("Tipos de licencia")
plt.xticks(rotation=45)
plt.show()
```

Gran parte de estos proyectos tienen licensias del MIT, otra pequeña parte tiene licencias con Noassertion y Apache 2.0

```python
# Repositorios forkeados
sns.countplot(data=github_data, x='is_fork')
plt.title("Repositorio fork VERDADERO/FALSO")
plt.show()
```

```python
# contar cuantas veces aparece cada valor de esta categoria binaria
github_data['is_fork'].value_counts()
```

Vemos que ninguno de estos proyectos es ninguna copia y luego fueron modificados, todos son de un unico dueño

```python
# ver cantidad de repositorios archivados
sns.countplot(data=github_data, x='archived')
plt.title("Repositorios que han sido archivdaos")
plt.show()
```

La gran mayoria de estos repositorios no estan archivados, pero otra pequena parte si, de 500 repositorios con mas estrellas solo 8 estan archivados

```python
github_data['archived'].value_counts()
```

```python
# relacion entre estrellas y forks
sns.regplot(data=github_data, x='stars', y='forks')
plt.title("Estrellas vs Forks")
plt.xlabel("Estrellas")
plt.ylabel("Forks")
plt.show()
```

Se ve una relacion entre mientras mas estrellas mayores van hacer los forks que habran

```python
# relacion entre estrellas y visualizaciones
sns.scatterplot(data=github_data, x='stars', y='watchers')
plt.title("Estrellas vs visualizaciones")
plt.xlabel("Estrellas")
plt.ylabel("Visualizaciones")
plt.show()
```

Mientras mas estrellas tiene un proyecto mayores seran las visualizaciones, esto claramente es una correlacion positiva

```python
# relacion entre estrellas y open issues
sns.scatterplot(data=github_data, x='stars', y='open_issues')
plt.title("Estrellas vs Open issues")
plt.xlabel("Estrellas")
plt.ylabel("Open Issues")
plt.show()
```

mientras mas grande es un proyecto menos errores/sugerencias tienen

```python
# relacion entre forks y visualizaciones
sns.scatterplot(data=github_data, x='forks', y='watchers')
plt.title("Forks vs visualizaciones")
plt.xlabel("Forks")
plt.ylabel("Visualizaciones")
plt.show()
```

Mientras mas forks tiene un repositorio, tieneden a tener mas visualizaciones

```python
# relacion entre estrellas y tamaño en kb por repositorio
sns.scatterplot(data=github_data, x='stars', y='size_kb')
plt.title("Estrellas vs tamaño kb")
plt.xlabel("Estrellas")
plt.ylabel("Tamaño kb")
plt.show()
```

Mientras mas estrellas tiene un repositorio no necesariamente tiene que ser muy pesado en kb

```python
# relacion entre estrellas y lenguajes utilizados
plt.figure(figsize=(15, 8))
sns.barplot(x=github_data['language'], y=github_data['stars'], errorbar=None)
plt.title("Language vs Estrellas")
plt.xlabel("Stars")
plt.ylabel("language")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.regplot(data=github_data, x='forks', y='open_issues')
plt.title("Forks vs Open issues")
plt.xlabel("Forks")
plt.ylabel("Open issues")
plt.show()
```
